# Matplotlib

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,360,50)
y1 = np.sin(np.radians(x))
y2 = np.cos(np.radians(x))

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.set_title('Sinus- und Kosinusfunktion')
ax.set_xlabel('Winkel [Grad]')
ax.set_ylabel('Funktionswert')
ax.set_xlim(0,360)
ax.set_ylim(-1.1,1.1)
ax.grid()

ax.plot(x,y1,label='Sinus')
ax.plot(x,y2,label='Kosinus')
ax.legend()

plt.show()
```

## Polarkoordinaten

```python
from math import pi 

x = np.linspace(0,2 * pi)
r = np.linspace(0,10)

fig = plt.figure()
ax = fig.add_subplot(1,1,1,polar=True)

ax.set_title('Polarkoordinaten', fontsize=18,color='blue')
ax.set_xlabel('$\phi$')
ax.set_xlabel('R')
ax.set_xlim(0,pi)
ax.set_ylim(-1.1,1.1)
ax.grid()

ax.plot(r,r,label='Sinus')
ax.plot(x,x,label='Kosinus')
ax.legend()

ax.plot('x', 'y', {1:2})

plt.show([1,1,2])
```
