# matplotlib Grundlagen

`matplotlib` verfolgt zwei Anätze zur Darstellung von Grafiken:

* Der objekt-orientierte Ansatz, in dem eine baumartige Struktur aufgebaut und dargestellt wird
* Der Einsatz der globalen `pyplot`-Struktur


```python
import matplotlib.pyplot as plt

%matplotlib inline
```

## Figures

Alle Grafiken werden in Figures erstellt. Eine `Figure` dient als Container-Objekt für die verschiedenen Grafikkomponenten, den `Artists`, vor allem des `Axes-Objekten.

Eine Figure wird mit Angabe der Abmessungen und Auflösung definiert, die unveränderbar sind:

## Axes 

Die wichtigsten `Artists` sind die `Axes`, die über zahlreiche Eigenschaften und (Grafik-)Methoden verfügen. Sie sind selbst wieder Container und können sogar andere `Figures` enthalten.

`Axes` werden mit ihren Abmessungen (_Rectangle_) mit Hilfe der `add_axes`-Methode von `Figure` erzeugt,. Eine `Figure` kann mehrere `Axes` enthalten, die mit verschiedenen _Layouts_ angeordnet werden können:

```python
fig1 = plt.figure(figsize=(8,6), facecolor='lightskyblue',layout='constrained')
fig1.suptitle('Figure.suptitle()')
fig1.add_subplot()

#print(fig.axes)
ax1 = fig1.axes[0]
print(ax1)
ax1.plot(range(6),range(1,7), c= 'r')
pass
```

```python
fig2 = plt.figure(figsize=(2, 2), facecolor='lightskyblue',
                 layout='constrained')
fig2.suptitle('Figure')
ax2 = fig2.add_subplot()
ax2.set_title('Axes', loc='left', fontstyle='oblique', fontsize='medium')
```

```python
fig3 = plt.figure(fig2)
plt.show()
```
