# MP01 - Diagramme

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Plot

### Einfache Plots mit Beschriftungen

```python
X = np.linspace(0,2 * np.pi,100)
Y1 = np.sin(X)
Y2 = np.cos(X)


plt.title('Einfache Plots')
plt.plot(X,Y1)
plt.plot(X,Y2)

plt.xlabel('Beschriftung der x-Achse')
plt.ylabel('Beschriftung der y-Achse')

plt.show()
```

### Plots mit zusätzlichen Features

```python
X = np.linspace(0,2 * np.pi,200)
Y = np.sin(X*X)

plt.title('Latex für Titel: $ sin(x^2)$')
plt.plot(X,Y,c='#ff0000cc',marker='x',markevery=10)

plt.xlabel('x')
plt.ylabel('$ sin(x^2)$')

plt.grid(True,lw=2,ls='--',c='.9')  # Background grid
plt.annotate('Maximum',ha = 'center',va = 'bottom',
             xytext=(1,0),xy=(1.25,1),
             arrowprops={'arrowstyle' : '->','facecolor' : 'green'})

plt.show()

```

## Scatter

### Einfache Scatterplots

```python
X = np.linspace(0,2 * np.pi,20)
Y = np.sin(X)
plt.title('Sinuskurve')
plt.scatter(X,Y)
plt.show()
```

### Komplexere Scatterplots

```python
X = np.linspace(0,2 * np.pi,20)
plt.title('Sinus- und Cosinuskurve')
Y1 = np.sin(X)
Y2 = np.cos(X)
plt.scatter(X,Y1,color=(1,0,0),edgecolor='k')
plt.scatter(X,Y2,marker='x',color=(0,1,0))

plt.show()
```

## 3D-Plot

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-3,3,256)
y = np.linspace(-3,3,256)

X,Y = np.meshgrid(x,y)
Z = np.sinc(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X,Y,Z,cmap = cm.jet)


plt.show()
```
