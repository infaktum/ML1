# PD01 Data Frames

```python
import pandas as pd
```

## Laden der CSV-Datei

```python
real_estates = pd.read_csv('RealEstate.csv', delimiter =',', index_col = "No")
real_estates.head()
```

## Umbenennen der Spalten

```python
real_estates.rename(columns = {'X1 transaction date': 'TransactionDate',
                               'X2 house age': 'HouseAge', 
                               'X3 distance to the nearest MRT station': 'NearestMrtStation', 
                               'X4 number of convenience stores': 'NumberOfStores',
                               'X5 latitude': 'Latitude',
                               'X6 longitude': 'Longitude',
                               'Y house price of unit area': 'HousePrice'
                              }, inplace=True)
real_estates.head()
```
