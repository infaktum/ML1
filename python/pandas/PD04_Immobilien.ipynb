# Weitere Features von Pandas

```python
import pandas as pd
```

## Laden der CSV-Datei

```python
immobilien = pd.read_csv('immobilien.csv', delimiter =',', index_col = "No")
immobilien.head()
```

## Umbenennen der Spalten

```python
immobilien.rename(columns = {'X1 transaction date': 'TransactionDate',
                               'X2 house age': 'HouseAge', 
                               'X3 distance to the nearest MRT station': 'NearestMrtStation', 
                               'X4 number of convenience stores': 'NumberOfStores',
                               'X5 latitude': 'Latitude',
                               'X6 longitude': 'Longitude',
                               'Y house price of unit area': 'HousePrice'
                              }, inplace=True)
immobilien.head()
```

## Informationen zu den Daten

```python
immobilien.info()
```

## Eine erste Statistik

```python
immobilien.describe()
```

## Geodaten

```python
immobilien.plot(kind='scatter', x = 'Longitude', y = 'Latitude');
```
