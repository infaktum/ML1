# Die Iris-DB - Analyse der Ergebnisse

Wir haben nun drei verschiedene Lösungen für unser Klassifizierungsproblem gefunden:

* Durch Visualisierung haben wir Grenzwerte für die vier Parameter _erraten_ und diese Werte fest implementiert. Wir haben dadurch schnell eine Trefferquote von über 92 % erreicht. Hier haben wir die Arbeit selbst erledigt.
* Mit der k-Means-Methode haben wir durch einen festen Algorithmus die Cluster bestimmt, der bei dem davon ausgegangen wird, dass sich die Werte um Mittelpunkte grupperen. Das ist unser Modell.
* Mit dem Neuronalen Netzwerk haben wir eine Verlustfunktion auf einer Auswahl von Trainngsdaten minimiert. Dabei war es notwendig, dass die Trainingsdaten bereits "getaggt" sind, d.h. die Iris-Sorte war bekannt.

```python
import numpy as np
import matplotlib.pyplot as plt

import knn as k 
import iris as ir
```

### Erstellung von Trainings- und Testdaten aus der Iris-Datenbank

```python
iris = ir.IrisDataSet("iris.csv")
trainings_daten, test_daten = iris.tt_daten(100)
```

Wir laden die Daten des trainierten Netzwerks:

```python
knn = k.NeuralNetwork(4,3,4)
knn.load("iris")
```

Wir setzen unsere drei _Mittelwerte_ aus dem k-Means-Einsatz in unser Netzwerk ein. Wir erwarten, dass diese drei Werte für "ideale" Vertreter der jeweiligen Iris-Sorte gelten und daher auch vom Netzwerk "besonders gut" erkannt werden.

```python
inputs =[ [5.006,3.418,1.464,0.244],[5.9016129,2.7483871,4.39354839,1.43387097],[6.85,3.07368421,5.74210526,2.07105263]]
outputs = [knn.query(i) for i in inputs]

plt.figure(figsize=(10, 4))
for col in [0,1,2]:
    plt.subplot(1,3,col+1)
    plt.bar([0,1,2], outputs[col], align='center') 
    plt.title(iris[col] )

plt.show()
```
