# Multi-Layered Perceptron und MNIST

## Laden des MNIST-Datensatzes

Der Datensatz ist recht groß und muss deshalb heruntergeladen werden. OpenML hat ihn vorrätig, und sklearn hat eine entsprechende Funktion:

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784')
X, y = mnist.data.to_numpy(), mnist.target.to_numpy()
```

## Visualisierung

```python
import matplotlib.pyplot as plt

def gallery(images, rows, cols, cmap='grey'):
    for n in range(rows * cols):          
        plt.subplot(rows, cols,n+1)
        ax = plt.gca()
        ax.axes.xaxis.set_visible(False)
        ax.axes.yaxis.set_visible(False)          
        plt.imshow(images[n], cmap) 
```

```python
images = [x.reshape(28,28) for x in X]

gallery(images,5,8,cmap='Blues') 
plt.show()    
```

## Der übliche Ablauf...

Nachdem die Daten vorbereitet sind, ist der weitere Ablaufwie üblich:

### Aufteilung der Daten

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

### Erzeugung und Training des Klassifizieres

```python
from sklearn.neural_network import MLPClassifier

mlpc = MLPClassifier(hidden_layer_sizes = 128, max_iter=100)

mlpc.fit(X_train, y_train)
```

## Wie gut arbeitet der Klassifizierer?

```python
score = mlpc.score(X_test, y_test)
print(f'Genauigkeit: {score:.1%}')
```

## Ein echter Anwendungsfall

Auch wenn der MNIST-Datensatz heute als Benchmark keine Rolle mehr spielt, zeigt er doch deutlich die Leistungsfähigkeit Neuronaler Netze auf. Ein genügend großer Datensatz vorausgesetzt, ermöglicht selbst ein einfaches Netzwerk wie das MLP mit kurzem Trainingsaufwand selbst auf einfachen Computern die Erstellung eines schon recht genauen Klassifizierers.

Trotz oder gerade wegen seiner Einfachheit stellt das MLP auch in den modernsten Anwendungen die Basis dar, eingeschlossen den GPT-LLMs. Es ist das Arbeitspferd der KI.
