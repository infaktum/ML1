# K-Means-Algorithmus

Ziel von k-Means ist es, einen Datensatz so in k Partitionen zu teilen, dass die Summe der quadrierten Abweichungen von den nächstgelegenen Cluster-Schwerpunkten minimal ist. Mathematisch entspricht dies der Optimierung der Funktion

$$  J = \sum_{i=1}^k  \sum_{x_j \epsilon S_i} || x_j - \mu_i||^2$$

mit den Datenpunkten $x_i$ und den Schwerpunkten $\mu_i$ der Cluster $S_i$. 

Diese Zielfunktion basiert auf der Methode der kleinsten Quadrate, und man spricht auch von **Clustering durch Varianzminimierung**, da die Summe der Varianzen der Cluster minimiert wird. Da zudem $|| x_j - \mu_i||^2$ die quadrierte Euklidische Distanz ist, ordnet k-Means effektiv jedes Objekt dem nächstgelegenen (nach Euklidischer Distanz) Clusterschwerpunkt zu. Umgekehrt ist das arithmetische Mittel ein Kleinste-Quadrate-Schätzer, optimiert also ebenfalls dieses Kriterium.

## Laden der Iris-Datenbank

```python
from sklearn.datasets import load_iris
iris_db = load_iris()

X, y, x_names, y_names = iris_db.data, iris_db.target, iris_db.feature_names, iris_db.target_names
print (f'Parameter:\t{x_names}\nTargets:\t{y_names}')
```

## Clusterbildung

```python
from sklearn.cluster import KMeans

Kmean = KMeans(n_clusters=3,verbose=0,max_iter=10,n_init = 'auto')
Kmean.fit(X)
cluster_centers = Kmean.cluster_centers_
labels = Kmean.labels_
```

## Visualisierung

```python
import matplotlib.pyplot as plt
plt.set_cmap('prism')
%matplotlib inline

fig = plt.figure( figsize=(12, 8), dpi=80)

ax_1, ax_2 = 0, 1


plt.scatter(X[:,ax_1], X[:,ax_2], s=100, marker='o',c=y)

for cc in cluster_centers:
    plt.scatter(cc[ax_1],cc[ax_2],s=100,marker='x')
    
plt.show();
```
