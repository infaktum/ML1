# Datenerzeugung mit `fetch`

Mit `Scikit-learn`lassen sich auch einige interessante Datensätze aus dem Internet laden, die zu groß sind, als dass man sie der Installation hätte beilegen können. Dies betrifft insbesondere größere Datensätze mit Bilddaten.

Die Datensätze können mit den `fetch*`-Funktionen des Pakets `sklearn.datasets` geladen werden, was wegen deren Größe schon einige Zeit dauern kann. Die heruntergeladenen Daten werden in einem Ordner auf der lokalen Festplatte abgelegt und beim nächsten Zugriff von dort eingelesen. Wie bei den `load`-Funktionen ist das Ergebnis wieder ein Objekt vom Typ `Bunch`.

### Lokaler Speicherort

Der lokale Speicherort kann durch die Umgebungsvariable __SCIKIT_LEARN_DATA__ gesetzt und mit `get_data_home()` abgefragt werden:

```python
import sklearn.datasets as ds

print(ds.get_data_home())
```

Dadurch soll verhindert werden, dass große Datensätze mehrfach geladen werden. Bei den `fetch`-Methoden kann das Verzeichnis explizit durch den Parameter `data_home` gesetzt werden.
