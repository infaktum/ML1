# Datenerzeugung mit `make`

Am flexibelsten ist die Erstellung zufälliger Daten mit den `make*`-Methoden aus dem paket `sklearn.datasets`. Für die mesiten Verfahren können hier passende Datensätze zu Entwicklung und Tests erstellt werden, wobei viele Steuerungs-Parameter gesetzt werden können. 

## Standardparameter

Die beiden wichtigsten Parameter sind

* __n_samples__: Die Anzahl der erzeugten Datensätze
* __n_features__: Die Anzahl der Features (Ausprägungen) je Datensatz

Jeder `make`-Methode hat Standardeinstellungen für die Parameter, so dass auch ein einfacher Aufruf ohne Parameter ein Ergebis liefert.

Ein weiterer praticher Parameter ist __random_state__, der den Zufallszahlen-Generator initialisiert. Mit ihm ist es möglich, die ansonsten (pseudo-)zufällig generierten Daten immer wieder reproduzieren zu können. Das ist für Entwicklungs- und testzwecke enorm praktisch, und auch wir werden dies hier immer wieder verwenden.


## Die erzeugten Daten

Das Ergebnis einer `make`-Methode ist jeweils ein `NumPy`-Array. Je nach Verfahren, für das die Methode gedacht ist, enthalten die Daten nur ein Array mit den Inputdaten __X__, oder zwei Arrays __X__, __y__ mit den Inputdaten und den Targetdaten. 

## Verschiedene Beispiele

Es stehen etwa 20 verschiedene Methoden zur Verfügung, von denen wir hier nur zwei vorstellen wollen. Für die Visualisierung der Daten verwenden wir `MatPlotLib`.

Zuerst importieren wir die benötigten Biblioheken:

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as ds
```

### Blobs (Daten für Clustering-Algorithmen

Bei Clustering-Algorithmen wie etwa __k_means__ untersuchen wir Daten, die sich in verschiedene Cluster aufteilen lassen. Mit `make_blobs` können wir geeignete Daten dazu erstellen. 

In unserem Beispiel erstellen wir 100 Testdaten mit zwei Features, die sich in fünf Cluster aufteilen lassen:

```python
data = ds.make_blobs(n_samples = 100,n_features = 2, centers = 5, return_centers = True, random_state = 42)
X, y, centers = data

plt.scatter(X[:,0],X[:,1], cmap = 'prism', c = y);
plt.scatter(centers[:,0], centers[:,1], marker = 'x');
```

### Regression

Im Gegensatz zur Clusterung wird bei der Regression eine stetige Funktion gesucht, d.h. die Menge __y__ der Targets stehen nicht für eine endliche Anzahl von Klassen, sondern können beliebige Werte sein.

Mit `make_regression` können Beispieldaten für Regressionen in verschiedenen Dimensionen erstellt werden. Dabei werden von einen linearen Abbildung $f:\mathbb{R}^M \rightarrow \mathbb{R}^N$ eine Anzahl Samples gezogen und verrauscht:

```python
data = ds.make_regression(n_samples = 100, n_features = 1, coef = True, noise= 10,random_state = 42)
X, y, coef = data

plt.scatter(X,y);
rng = [np.min(X), np.max(X)]
plt.plot(rng,coef * rng);
```
