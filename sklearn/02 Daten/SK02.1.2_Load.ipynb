# Datenerzeugung mit `load`

`Scikit-learn` kommt mit etlichen "eingebauten" Datensätzen daher, die direkt benutzt werden können. Es handelt sich dabei um bekannte Datensätze aus der KI, wie etwa den __Iris-Datensatz__. Diese Beispieldaten sind nicht besonders umfangreich und vor allem zu Lernzwecken hilfreich. 

Die Datensätze können mit den `load*`-Funktionen des Pakets `sklearn.datasets` geladen werden. Die Daten werden nicht aus dem internet geladen, sondern liegen der `Scikit-learn`-Installation bei. Im Gegensatz zu den `make`-Methoden erhält man keine `Numpy`-Arrays zurück, sondern eine `sklearn`-eigene Struktur namens `Bunch`.

## Bunches

`Bunches` enthalten neben den Daten noch zusätzliche Meta-Informationen, die wie bei einem `Dictionary` über Keys abgerufen werden können. Tatsächich ist `Bunch`ist eine Unterklasse von `Dictionary`mit dem Vorteil, dass die Werte direkt als Attribute zur Verfügung stehen. Wir werden dies gleich in den beispielen demonstrieren.

## Beispiele


Wir laden zunächste die benötigten Pakete:

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as ds

from dl1 import gallery
```

### Der Iris-Datensatz

Die __Iris-Datenbank__ ist eines der bekanntesten und beliebtesten Beispiele für den Einsatz von KI im Bereich der __Klassifizierung__. Der Iris-Datensatz besteht aus insgesamt 150 Datenzeilen mit jeweils vier Einträgen (__Features__) für die Breite und Höhe (in cm) zweier Blätter der Iris-Blüten. Die 150 Datenzeilen sind dabei in drei Gruppen (__Klassen__) zu je 50 Beispielen aufgeteilt:

* Iris setosa
* Iris virginica
* Iris versicolor

Wir laden die Datenbank und lassen uns die Schlüssel des erhaltenen `Bunches` anzeigen:

```python
iris = ds.load_iris()
print(f'Keys: {iris.keys()}')
```

Über den Schlüssel __DESCR__ erhalten wir wie bei allen Bunches Meta-Informationen über den Datensatz. Wir lassen uns einen Teil davon anzeigen:

```python
print(iris.DESCR[:500])
```

_data_ und _target_ enthalten die Werte für __X__ bzw. __y__. _target_names_ enthält die zu __y__ gehörenden Namen der Irisblüten:

```python
X, y, names = iris.data, iris.target, iris.target_names
print(names)
```

Man kann die Daten auch ohne Umwege über einen `Bunch` laden:

```python
X, y = ds.load_iris(as_frame = True, return_X_y = True)
print(X[:5])
```
