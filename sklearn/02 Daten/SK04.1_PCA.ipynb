## Beispiel: Der Iris-Datensatz


Wir führen die Arbeitsweise der PCA am __Iris__-Datensatz vor:

```python
from sklearn.datasets import load_iris

iris = load_iris()
X, y = iris.data, iris.target
```

```python
print(iris.target_names)
print(X[::15])
```

Der Iris-Datensatz besitzt vier Features - die Abmessungen zweier Blättersorten. Wir erstellen hier einmal eine mögliches zweidimensionale Projektion - die auf die Abmessungen des Petal-Blattes:

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 2],X[:,3], cmap = 'prism', c = y);
```

Der roten Punkte stehen für die Iris setosa, die sich sehr leicht anhand der Länge und Breite des Petal-Blattes bestimmen läasst.

Wir wollen versuchen, die Zahl der Features auf zwei zu reduzieren. Dies macht Sinn, da die Länge und Breite eines Blattes (Petal und Sepal) korrelieren sollten, da seine Form unabhängig von seiner Größe ist.

Wir finden das PCA-Objekt im Modul `sklearn.decomposition`:

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)
```

So sehen die tranformierten Daten aus:

```python
print(iris.target_names)
print(X_pca[::15])
```

Die ersten vier Einträge gehören zur Iris setosa, die folgenden drei jeweils zu Iris versicolor und Iris virginica. Sie heben sich deutlich voneinander ab.

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_pca[:, 0],X_pca[:, 1], cmap = 'prism', c = y);
```

Wir haben nun nur noch zwei Features statt der originalen vier. Die Grafik zeigt uns, dass wir "nichts verloren" haben. Die Sorte Iris setosa (rote Punkte) ist immer noch völlig von den beiden anderen Sorten getrennt. Grün und Gelb (Iris virginica und Iris versicolor) überlappen sich nach wie vor nur in einem sehr kleinen Bereich.
