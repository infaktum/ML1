## Statistik

Wir können NumPy-Arrays als mathematische Objekte auffassen, nämlich als Vektoren (oder Tensoren) der Linearen Algebra.

Aber natürlich kann man diese Arrays auch nur als eine Zusammenfassung von Zahlen auffassen, wie sie bei Datensammlungen auftraten. Messen wir etwa eine Größe 1000 Mal, so erhlten wir 1000 Werte, die wir in einem Array zusammenfassen können. Haben wir bei jeder Messung mehrere Werte (z.B. Länge, Breite, Höhe oder Körpergröße und Gewicht), so haben diese Arrays nicht nur mehrere Zeilen, sondern auch Spalten. Hier macht es keinen Sinn mehr, diese Objekte miteinander zu multiplizieren. Wir können aber statistische Werte wir Mittelwerte berechnen. Auch dies geht mit NumPy sehr einfach.

### Mittelwerte

Wir erzeugen normalverteilte Zufallswerte und stellen sie mit matplotlib in einem Histogramm dar:


```python
X = np.random.normal(0,100,(10000))

plt.hist(X, bins='auto');
```

### Mittelwert, Varianz und Standardableitung

Die wichtigsten Werte wie Mittelwert und Standardabweichung lassen sich leicht berechnen:

```python
mittelwert = np.sum(X) / X.size
varianz = np.sum((X - mittelwert) ** 2) / X.size
std_abweichung = np.sqrt(varianz)

print(f'Mittelwert: \t{mittelwert}\nVarianz: \t{varianz}\nStdAbweichung: \t{std_abweichung}')
```

Für diese Werte stellt Numpy auch direkt Funktionen zur Verfügung:

```python
mean, var, std_dev = np.mean(X), np.var(X),np.std(X)
print(f'Mittelwert: \t{mittelwert}\nVarianz: \t{varianz}\nStdAbweichung: \t{std_abweichung}')
```

### Definition der Normalverteilung

```python
def normal_verteilung(mu,sigma):
    return  lambda x: (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2)) 
```

#### Darstellung der Standard-Normalverteilung

```python
mu, sigma = 0,1

std_nv = normal_verteilung(mu, sigma)

xval = mu + np.array([0,-sigma,sigma])
yval = std_nv(xval)
X = np.linspace(mu-3*sigma,mu + 3*sigma,500)

plt.grid(True)
plt.scatter(xval,yval,c='r')
plt.plot(X,std_nv(X));
```

### Korrelation

```python
X = np.random.randint(0,100,(10000))
Y = np.random.randint(0,100,(10000))
```

```python
Xs = (X - (np.mean(X))) /  np.std(X)
Ys = (Y - (np.mean(Y))) /  np.std(Y)
```

```python
cov = np.cov(Xs,Ys)
print(cov)
```

```python
np.linalg.det(cov)
```
