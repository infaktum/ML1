# Regression

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.array(list (range(5)))

y = np.array([ 5,  9, 10, 14, 20])
```

```python
plt.scatter(x,y)
plt.plot(x,y,c="r")
plt.show();
```

```python
def gerade(m,b):
    return lambda x: m*x + b
    
def D(m,b,x,y):
    g = gerade(m,b)
    return np.sum((g(x) - y) ** 2 )
```

```python
print(D(3,5,x,3 * x +5))
```

```python
def koeffizienten(x,y):
    sum_x = np.sum(x)
    sum_xx = np.sum(x*x)
    sum_y = np.sum(y)
    sum_xy = np.sum(x*y)
    N = len(x)
    return sum_xx, sum_x, sum_xy, sum_x, N, sum_y

def solve (a1,b1,c1,a2,b2,c2):
    l2 = (a2 * c1 - a1*c2) / (a2 * b1 - a1 * b2)
    l1 = (c1 - b1 * b2) / a1
    return l1,l2
def lin_reg(x,y):
    a1,b1,c1,a2,b2,c2 = koeffizienten(x,y)
    m,b = solve (a1,b1,c1,a2,b2,c2)
    return m,b

print (lin_reg(x,3*x + 5))
```

```python
m,b = lin_reg(x,y)
plt.scatter(x,y)
plt.plot(x,m*x +b)
plt.show();
```
