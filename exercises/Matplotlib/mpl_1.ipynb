# Übung zu matplotlib: Punkte, Plots

```python
from random import randint
import matplotlib.pyplot as plt
```

Erstellen Sie zwei Listen **x** und **y**. Dabei soll **x** die ganzen Zahlen von 1 bis 20 enthalten, und y eine gleichgroße Liste von Zufallszahlen zwischen -5 und 5. Eine solche Zufallszahl erhalten Sie mit **randint(-5,6)**.

```python
N = 20
x = list(range(1,N))
y = [randint(-5,6) for _ in x]
```

### Punkte-Diagramm

Zeichnen Sie die Wertepaare **x** und **y** in einem Punkte-Diagramm (*Scatter Diagram*)

```python
plt.scatter(x,y);
```

### Balken-Diagramm

```python
plt.bar(x,y);
```

### Funktions-Plot

Zeichnen Sie nun eine verbindende Kurve (die eckig sein wird) durch die Punkte, wie in einem Funktions-Diagramm. Zeichnen Sie zusätzlich die Punktepaare wie im Punkte-Diagramm ein. 

Können Sie die Punkte auch mit einer anderen Farbe zeichnen, etwa Rot? Das geht mit dem Zusatzparameter **c='red'**.  Färben Sie die Kurve dann in einer Farbe Ihrer Wahl.

```python
plt.scatter(x,y,c='red')
plt.plot(x,y,c='g')
plt.show()
```

```python
def f(N):
    x = list(range(-1,N))
    y =  [randint(-5,6) for _ in x]
    return x,y
```
