# Übung: Lineare Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Die Situation

Wir haben eine Menge von Punktepaaren vorliegen, und zwar als zwei Listen von Daten **x** und **y** mit jeweils **N**=10 Punkten:

```python
N = 10
x = list(range(N))
y = [3,7,7,8,10,12,11,15,14,17]
```

### Visualisierung

Gibt es einen **funktionalen Zusammenhang** zwischen den Werten? Wir wollen uns einen Überblick über die  verschaffen!

Visualisieren Sie die Daten in einem Punkte-Diagramm mit Hilfe der Funktion **scatter** von **matplotlib**:

```python
plt.scatter(x,y);
```

Mit etwas Phantasie liegen die Punkte auf einer Geraden. Dies ist unser **Modell**: Es gibt einen linearen Zusammenhang zwischen den **x** und **y**.

## Bestimmung der Ausgleichgeraden (Lineare Regression)

Wir suchen die Gerade, die die **Fehlerquadratsumme** der Abstände minimiert. Eine Gerade **g** wird durch zwei **Parameter** definiert: ihre Steigung **m** und den Ordinatenabschnitt **b**: $$ g(x) = m x + b$$

Schreiben Sie eine Funktion, die zu zwei Parametern **m** und **b** die Gerade **g** als Python-Funktion zurückliefert. Dies geht zum Beispiel mit einem Lambda-Ausdruck:

```python
def gerade(m,b):
    def g(x):
        return  m * x + b
    return g
```

```python
x = np.array(x)
g = gerade(2,3)
print(g(x))
```

```python
plt.plot(x, g(x));
```
